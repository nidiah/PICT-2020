Los nombres de archivos son muy asistemáticos, es un quilombo. Extraer todo y después normalizar los nombres de archivos sacando todos los guiones y el AG

```python
nuevas_niv1
```

```python
def norm_filename(s):
    return s.rsplit('_', 1)[0]#.replace('AG', '').replace('_', '')

nuevas_niv1['Archivo'] = nuevas_niv1['Recorte'].map(norm_filename)
nuevas_niv2['Archivo'] = nuevas_niv2['Recorte'].map(norm_filename)

def norm_recorte1(s):
    return s.replace('_', '', 5).replace('AG', '')

def norm_recorte2(s):
    return s.replace('_', '', 3).replace('AG', '')

nuevas_niv1['Recorte'] = nuevas_niv1['Recorte'].map(norm_recorte1)
nuevas_niv2['Recorte'] = nuevas_niv2['Recorte'].map(norm_recorte2)
```

```python
nuevas_niv2#['Recorte'].map(norm_recorte1)
```

```python
def find_year1(s):
    return s.split('_')[0]

informantes = {
    'AC': 'Andres Crespo',
    'AJ': 'Agustín Juárez',
    'CS': 'Celestina Sacarías',
    'E': 'Ernestina',
    'ER': 'Elizabeth Rojas',
    'FR': 'Félix Ramírez',
    'FAR': 'Faustino Ramírez',
    'GR': 'Graciano Ramírez',
    'JR': 'José Rojas',
    'MV': 'Mauricio Valdez',
    'SD': 'Sofía Díaz', #'Chula'    
    'SR': 'Sara Rojas', #'Sara'
    'TS': 'Teresita Sánchez',
    'TeS': 'Teo Servín'
}

def find_spk1(s):
    return informantes[s.split('_')[3]]
    
nuevas_niv1['Código'] = 'Gu'+nuevas_niv1['Archivo'].map(find_year1)
nuevas_niv1['Cita Audio'] = "Gutiérrez, "+nuevas_niv1['Archivo'].map(find_year1)
nuevas_niv1['Consultante'] = nuevas_niv1['Archivo'].map(find_spk1)

nuevas_niv1
```
