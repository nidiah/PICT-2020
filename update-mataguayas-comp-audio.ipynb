```python
import pandas as pd
```

```python
## SPREADSHEET
gsheet = "https://docs.google.com/spreadsheets/d/1GtliM_yutOBnWaferOmQjVNLt_xOxz56rst6ZR9MkXc/export?gid=1308046217&format=csv" ## CAICYT
gsheet = "palabras_en_comun_audios.csv" ## BELERIAND

## OUT
ghdir = "../pict-glotales/_data" ## Repertorio del sitio Github
```

```python
comun_audio =  pd.read_csv(gsheet)
comun_audio.sample(5)
```

Sacamos espacios en blanco al final del nombre del archivo de recorte de audio que se sube al sitio:

```python
comun_audio['Recorte'] = comun_audio['Recorte'].str.strip()
```

Los datos del cuadro de mataguayas comparadas están organizados así:

```python
mataguayas_comp_audio = pd.read_csv(f"{ghdir}/mataguayas-comparadas-audio.csv")
mataguayas_comp_audio.sample(3)
```

```python
len(mataguayas_comp_audio)
```

Seleccionamos las columnas que necesitamos de df de datos colaborativos:

```python
comun_audio = comun_audio[['Español','Lengua' ,'Palabra','Cita Audio', 'Recorte']]
comun_audio.sample(5)
```

Tenemos que ir lengua por lengua. Empezamos añadiendo las palabras en chorote:

```python
chorote = comun_audio[comun_audio['Lengua'] == 'chorote']
chorote.drop(columns='Lengua', inplace=True)
chorote = chorote.rename(columns={"Español":"espanol", "Cita Audio":"fuente_chorote", "Palabra":"chorote", "Recorte":"audio_chorote"})
print(f"Tenemos {len(chorote)} palabras en chorote para el cuadro de audio")
chorote.sample(5)
```

Ahora recuperamos las palabras en maka y las agregamos al nuevo df:

```python
maka = comun_audio[comun_audio['Lengua'] == 'maka']
maka.drop(columns='Lengua', inplace=True)
maka = maka.rename(columns={"Español":"espanol", "Cita Audio":"fuente_maka", "Palabra":"maka", "Recorte":"audio_maka"})
print(f"Tenemos {len(maka)} palabras maka para el cuadro de audio")
maka.sample(5)
```

```python
new_mataguayas_comp = chorote.merge(maka, how="outer", on="espanol")
new_mataguayas_comp#[30:40]
```

```python
new_mataguayas_comp = new_mataguayas_comp.drop(3)
```

Ahora añadimos datos nivacle:

```python
nivacle = comun_audio[comun_audio['Lengua'] == 'nivacle']
nivacle.drop(columns='Lengua', inplace=True)
nivacle = nivacle.rename(columns={"Español":"espanol", "Cita Audio":"fuente_nivacle", "Palabra":"nivacle", "Recorte":"audio_nivacle"})
print(f"Tenemos {len(nivacle)} palabras nivacle para el cuadro de audio")
nivacle.sample(5)
```

```python
new_mataguayas_comp = new_mataguayas_comp.merge(nivacle, how="outer", on="espanol")
new_mataguayas_comp
```

Ahora los datos de wichi:

```python
wichi = comun_audio[comun_audio['Lengua'] == 'wichi']
wichi.drop(columns='Lengua', inplace=True)
wichi = wichi.rename(columns={"Español":"espanol", "Cita Audio":"fuente_wichi", "Palabra":"wichi", "Recorte":"audio_wichi"})
print(f"Tenemos {len(wichi)} palabras wichi para el cuadro de audio")
wichi.sample(4)
```

```python
new_mataguayas_comp = new_mataguayas_comp.merge(wichi, how="outer", on="espanol")
new_mataguayas_comp
```

```python
new_mataguayas_comp = new_mataguayas_comp.sort_values(by="espanol")
new_mataguayas_comp
```

Guardamos en el repo del proyecto:

```python
new_mataguayas_comp.to_csv(f"{ghdir}/mataguayas-comparadas-audio.csv", index=False)
```
