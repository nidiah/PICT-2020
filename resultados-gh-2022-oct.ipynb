## 1. Modificar cuadro de comparación de fuentes textuales

```python
comp_texto = pd.read_csv(f"{wdir}/mataguayas-comparadas-AG.csv")
```

```python
comp_texto
```

```python
textfiles_info = pd.read_csv(f"{wdir}/_data/textfiles_info.csv")
```

```python
textfiles_info
```

```python
#textfiles_info.drop([12, 13], inplace=True)
textfiles_info[["Codigo", "Cita"]].sort_values(by=["Codigo"])
```

```python
for lang in langs:
    #print(lang.upper())   
    for i, row in comp_texto.dropna(subset=["fuente_wichi", "fuente_chorote",'fuente_nivacle', 'fuente_maka']).iterrows():
        for _, mapping_row in textfiles_info.iterrows():
            if not pd.isna(row[f"fuente_{lang}"]):
                if mapping_row["Codigo"] in row[f"fuente_{lang}"]:
                    #print(row[f"fuente_{lang}"])#, mapping_row["Cita"])
                    #print(row[f"fuente_{lang}"].replace(mapping_row["Codigo"], mapping_row["Cita"]))
                    
                    #row[f"fuente_{lang}"] = row[f"fuente_{lang}"].replace(mapping_row["Codigo"], mapping_row["Cita"])
                    comp_texto.loc[i][f"fuente_{lang}"] = row[f"fuente_{lang}"].replace(mapping_row["Codigo"], mapping_row["Cita"])
```

```python
comp_texto
```

```python
comp_texto.to_csv(f"{wdir}/mataguayas-comparadas-AG.csv", index=False)
```
