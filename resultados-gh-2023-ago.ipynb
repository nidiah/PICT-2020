```python
tcfil = pd.read_excel(f"{tcfildir}/TCFiladelfia.xlsx")
tcfil.head()
```

```python
informantes = {
    'AC': 'Andres Crespo',
    'AJ': 'Agustín Juárez',
    'CS': 'Celestina Sacarías',
    'E': 'Ernestina',
    'ER': 'Elizabeth Rojas',
    'FR': 'Félix Ramírez',
    'FAR': 'Faustino Ramírez',
    'GR': 'Graciano Ramírez',
    'JR': 'José Rojas',
    'MV': 'Mauricio Valdez',
    'SD': 'Sofía Díaz', #'Chula'    
    'SR': 'Sara Rojas', #'Sara'
    'TS': 'Teresita Sánchez',
    'TeS': 'Teo Servín'
}
```

```python
##
## EN VEZ DE HQCER TODO ESTO, HACER UN MERGE
##

recortes_dir = '../pict-glotales/assets/audio'

new_rows = []

for _, row_sel in sel_niv.iterrows():
    for _, row_tcfil in tcfil.iterrows():
        if row_sel['Español'] == row_tcfil['Glosa']:
            
            #print(row_sel['Español'], row_tcfil['AFI'])
            
            row_sel['Palabra'] = row_tcfil['AFI']
            row_sel['ID Audio'] = row_tcfil['ID']
            row_sel['Código'] = f"Gu{row_tcfil['ID'].split('_')[0]}"
            row_sel['Cita Audio'] = f"Gutiérrez_{row_tcfil['ID'].split('_')[0]}"
            id_consultante = row_tcfil['ID'].split('_')[3]
            row_sel['Consultante'] = informantes[id_consultante]
            row_sel['Original_audiopath'] = row_tcfil['Audiofile']
            
            ## nombredelarchivofuenteresumido_glosaenespañol. Ej. 20120318_003_algarrobo
            nombre_recorte = row_tcfil['ID'].replace("_", "").replace("AG", "")+"_"+row_tcfil['Glosa'].replace(" ", "_")
            row_sel['Recorte_path'] = f"{recortes_dir}/{nombre_recorte}.wav"
            
        
#         ## Haya coincidencia o no, agrego la fila porque preservo las palabras para 
#         ## las que no se encontró nada en TCFiladeflia
        new_rows.append(row_sel)
```

### 3.1 Chorote

```python
gsheet = "https://docs.google.com/spreadsheets/d/1GtliM_yutOBnWaferOmQjVNLt_xOxz56rst6ZR9MkXc/export?gid=1308046217&format=csv" ## CAICYT
#gsheet = "palabras_en_comun_audios.csv" ## BELERIAND

comun_audio =  pd.read_csv(gsheet)
```

Tenemos que agregar las palabras de la tanda 2023-08 (columna "Tanda")

```python
nuevas_chorote = comun_audio[ (comun_audio['Tanda'] == '2023-08' ) & (comun_audio['Lengua'] == 'chorote')]

print(f"Tenemos {len(nuevas_chorote)} palabras nuevas en chorote para agregar al cuadro")
```

Seleccionamos las columnas que necesitamos y las renombramos para poder hacer merge:

```python
nuevas_chorote = nuevas_chorote[['Español', 'Palabra','Cita Audio', 'Recorte']] ## Cambiaron el nombre de la columna 'Cita' por 'Cita Audio'
nuevas_chorote = nuevas_chorote.rename(columns={"Español":"espanol", "Cita Audio":"fuente_chorote", "Palabra":"chorote", "Recorte":"audio_chorote"})
```

```python
## Mostramos las palabras chorte que vamos a agregar
nuevas_chorote#.head()
```

Ahora cargamos los datos del cuadro de mataguayas comparadas por fuente de audio:

```python
mataguayas_comp_audio = pd.read_csv(f"{ghdir}/mataguayas-comparadas-audio.csv")
mataguayas_comp_audio
```

```python
print(f"El cuadro comparativo actual tiene {len(mataguayas_comp_audio)} palabras")
```

Hacemos la fusión de los datos y mostramos:

```python
merge = mataguayas_comp_audio.merge(nuevas_chorote, how="outer")
merge[30:40]
```

⚠️⚠️⚠️

Detectamos que en el caso del chorote, hay muchas palabras duplicadas. A veces cambia la fuente (por ej, 4 y 5), a veces cambia la transcripción (41 y 52) a veces coincide la fuente y la transcripción pero son distintas muestras (por ej, 33 y 58):

⚠️⚠️⚠️

```python
duplicadas_chorote = merge[merge.sort_values(by=['espanol', 'chorote']).duplicated(subset=['espanol'], keep=False)].sort_values(by='espanol')
print(f"Hay {len(duplicadas_chorote)} palabras duplicadas en chorote")
duplicadas_chorote[['espanol', 'chorote', 'fuente_chorote', 'audio_chorote']] ## mostramos sólo la info para el chorote
```

¿Queremos palabras repetidas en una misma lengua pero de diferentes fuentes?
❓❓❓❓

```python
duplicadas_chorote = duplicadas_chorote[['espanol', 'chorote', 'fuente_chorote', 'audio_chorote']]
duplicadas_chorote.to_excel('duplicadas_chorote_cuadro_mataguayas_audio.xlsx', index=False)
```
