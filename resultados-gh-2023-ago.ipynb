## 1. Encontrar palabras seleccionadas en TCFiladelfia y llenar info en df de selección de recortes

```python
## Selección de palabras hecha por AG:
gsheet = "https://docs.google.com/spreadsheets/d/1resasym4ZmqI7xytQknT8Veeh_1tBgOfOU_mwCQ2EVE/export?gid=0&format=csv"

sel_niv = pd.read_csv(gsheet)
```

```python
tcfil = pd.read_excel(f"{tcfildir}/TCFiladelfia.xlsx")
tcfil.head()
```

```python
informantes = {
    'AC': 'Andres Crespo',
    'AJ': 'Agustín Juárez',
    'CS': 'Celestina Sacarías',
    'E': 'Ernestina',
    'ER': 'Elizabeth Rojas',
    'FR': 'Félix Ramírez',
    'FAR': 'Faustino Ramírez',
    'GR': 'Graciano Ramírez',
    'JR': 'José Rojas',
    'MV': 'Mauricio Valdez',
    'SD': 'Sofía Díaz', #'Chula'    
    'SR': 'Sara Rojas', #'Sara'
    'TS': 'Teresita Sánchez',
    'TeS': 'Teo Servín'
}
```

```python
recortes_dir = '/home/nidia/DILA/pict-glotales/assets/audio'

new_rows = []


for _, row_sel in sel_niv.iterrows():
    for _, row_tcfil in tcfil.iterrows():
        if row_sel['Español'] == row_tcfil['Glosa']:
            
            row_sel['Palabra'] = row_tcfil['AFI']
            row_sel['ID Audio'] = row_tcfil['ID']
            row_sel['Código'] = f"Gu{row_tcfil['ID'].split('_')[0]}"
            row_sel['Cita Audio'] = f"Gutiérrez_{row_tcfil['ID'].split('_')[0]}"
            id_consultante = row_tcfil['ID'].split('_')[3]
            row_sel['Consultante'] = informantes[id_consultante]
            row_sel['Original_audiopath'] = row_tcfil['Audiofile']
            
            ## nombredelarchivofuenteresumido_glosaenespañol. Ej. 20120318_003_algarrobo
            nombre_recorte = row_tcfil['ID'].replace("_", "").replace("AG", "")+"_"+row_tcfil['Glosa'].replace(" ", "_")
            row_sel['Recorte_path'] = f"{recortes_dir}/{nombre_recorte}.wav"
            
        #else:
            #no_encontradas.update(row_sel['Español']) ## esto genera conjunto de letras,mejorar
        
        ## Haya coincidencia o no, agrego la fila porque preservo las palabras para 
        ## las que no se encontró nada en TCFiladeflia
        new_rows.append(row_sel)
```

```python
df = pd.DataFrame(new_rows)
df = df.drop_duplicates()
df
```
