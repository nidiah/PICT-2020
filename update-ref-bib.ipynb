```python
import pandas as pd
```

Recueperamos la información bibliográfica corregida por AG en `glotales_geoloc`:

```python
sheet = "https://docs.google.com/spreadsheets/d/1q_HUGbLVsyo6P6ue9J26GH4UBmwz1G8b/export?1861121093&format=csv" ## CAICYT
#sheet = "ref_biblio.csv" ## BELERIAND

df = pd.read_csv(sheet)
df.head(5)
```

Nos quedamos sólo con las columnas pertinentes para la información de fuentes (la info de lengua la recuperamos de otra planilla porque acá está supeditada a los lugares donde se hizo el trabajo de campo).

```python
df = df[['Cita', 'Cita Bibliográfica', 'Soporte']]
df.rename(columns={'Cita Bibliográfica': 'Referencia'}, inplace=True)
df.head(5)
```

```python
df = df.drop_duplicates().sort_values(by='Cita')
df
```

Recuperamos información de lengua y de tipo de fuente de [map_archivos_ref_biblio](https://docs.google.com/spreadsheets/d/1b3AE-5y08rRTI5lmcd9ykr0JDwYlGEgtAcAOR5w572w/edit#gid=1411727896):

```python
sheet= "https://docs.google.com/spreadsheets/d/1b3AE-5y08rRTI5lmcd9ykr0JDwYlGEgtAcAOR5w572w/export?1411727896&format=csv"
df2 = pd.read_csv(sheet)
df2.head(5)
```

```python
df2 = df2[['Cita','Lengua', 'Tipo']].drop_duplicates().sort_values(by=['Cita', 'Lengua'])
df2
```

Juntamos todos los datos de Bucca, 1969 en uno solo porque originalmente era un sooo documento:

```python
df2.iat[0, 1] = 'chorote, nivaclé, wichí'
```

```python
df2 = df2.drop([18, 64])
df2
```

```python
bib = df.merge(df2).drop_duplicates().sort_values(by=['Cita', 'Lengua'])
bib
```

```python
bib.to_csv('../pict-glotales/_data/ref_biblio.csv', index=False)
```
